```python
import pandas as pd

df = pd.read_csv('cleaned_cuisines.csv')
df.head()
```

```python
cuisines_label_df = df['cuisine']
cuisines_label_df.head()
```

```python
df.cuisine.unique()
```

```python
cuisine_feature_df = df.drop(['Unnamed: 0', 'cuisine'], axis=1)
cuisine_feature_df.head()
```

```python
#importing the necessary libraries.

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report, precision_recall_curve
import numpy as np
```

```python
#splitting training and test data

X_train, X_test, y_train, X_test = train_test_split(cuisine_feature_df, cuisines_label_df, test_size=0.3)
```

```python
# LINEAR SVC CLASSIFIER(Support- Vector clustering)

#The C parameter refers to 'regularizaton'which regulates the influence of parameters.
#Kernel is set to linear to ensure leverage linear SVC
#Probability defaults to false, here we set it to true to gather probability
#random state is set to '0'to shuffle data to get probabilities

C = 10

# creating different classifiers

classifiers = {
    'Linear SVC': SVC(kernel='linear', C=C, probability=True, random_state=0), 'KNN classifier': KNeighborsClassifier(C)
}

```

```python
#training the model SVC and print out a report
n_classifiers = len(classifiers)

for index, (name,classifier) in enumerate(classifiers.items()):
    classifier.fit(X_train, np.ravel(y_train))

    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print('Accuracy (train) for %s: %0.1f%%' % (name, accuracy*100))
    print(classification_report(y_test, y_pred))

```
