OBJECTS AND CLASSES.
OBJECTS: PEOPLE
ATTRIBUTES: NAME, AGE, RACE, GENDER, WEIGHT, HEIGHT
METHODS: INTRODUCE SELF.

```python
class People:
    def __init__(self, name, age, race, gender, weight, height):
        self.name = name
        self.age = age
        self.race = race
        self.gender = gender
        self.weight = weight
        self.height = height

    def introduce_self(self):
        print(f'Hey there, my name is {self.name}, {self.age} years old, {self.race} in color, {self.gender}, {self.weight} kgs heavy and {self.height} cm tall.')
        

class Robots:
    def __init__(self, name, color, weight):
        self.name = name
        self.color = color
        self.weight = weight
    def introduce_self(self):
        print('My name is ', self.name)
        print('I am ', self.color)
        print('I weigh '+ str(self.weight)+ ' KGS')
```

```python
p1 = People('Peter Muhumuki', 18, 'Black', 'Male', 55, 119)
p2 = People('Alice', 20, 'White', 'Female', 70, 150)

r1 = Robots('Tom', 'Red', 40)
r2 = Robots('Jerry', 'Blue', 40)
```

```python
p1.introduce_self()
```
